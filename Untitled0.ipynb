```python
sleep_map = {
    'Very Poor': 1,
    'Poor': 2,
    'Average': 3,
    'Good': 4,
    'Very Good': 5
}

df['sleep_quality_num'] = df['sleep_quality'].map(sleep_map)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='daily_gaming_hours', y='academic_work_performance')
plt.title('Daily Gaming Hours vs Academic/Work Performance')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='daily_gaming_hours', y='work_productivity_score')
plt.title('Daily Gaming Hours vs Work Productivity')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sleep_quality_num', y='academic_work_performance')
plt.title('Sleep Quality vs Academic/Work Performance')
plt.show()
```

```python
df['academic_work_performance'].unique()
```

```python
df['work_productivity_score'].unique()
```

```python
performance_map = {
    'Poor': 1,
    'Below Average': 2,
    'Average': 3,
    'Above Average': 4,
    'Excellent': 5
}

df['academic_work_performance_num'] = df['academic_work_performance'].map(performance_map)
```

```python
productivity_map = {
    'Poor': 1,
    'Below Average': 2,
    'Average': 3,
    'Above Average': 4,
    'Excellent': 5
}

df['work_productivity_score_num'] = df['work_productivity_score'].map(productivity_map)
```

```python
corr = df[['daily_gaming_hours',
           'sleep_quality_num',
           'academic_work_performance_num',
           'work_productivity_score_num']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
